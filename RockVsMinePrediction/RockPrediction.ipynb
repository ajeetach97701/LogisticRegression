```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('SonarData.csv', header=None)
```

```python
df.head()
```

```python
df[60].value_counts()
```

```python
sns.countplot(data = df, x=df[60].value_counts())
```

```python
df.groupby(60).mean()
```

```python
X = df.drop(columns=60, axis=1)
y = df[60]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, stratify=y, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_hat = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('The accuracy obtained is',accuracy_score(y_test, y_hat))
```

```python
from sklearn.model_selection import cross_val_score
grid = cross_val_score(model, cv=20, X=X_train, y= y_train, scoring='accuracy')
```

```python
print("Accuracy for folds:",grid)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'penalty': ['l1','l2','elasticnet'],
    'C': [0.1, 0.01, 5, 10, 50, 100],
    'solver': ['saga', 'liblinear', 'lbfgs','newton-cg'], 
    'l1_ratio': [0, 0.05, 0.01, 0.1 , 1]
}

cv =[1,4,5,10,20,30,40]
clf = GridSearchCV(model,cv= 20, param_grid=param_grid)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_estimator_)
print(clf.best_params_)
```

```python
y_hat = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_hat)
```
